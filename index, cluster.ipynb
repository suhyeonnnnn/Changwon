# 클러스터용 파일 만들기

```python
df.shape
```

```python
#신호등이 있는 좌표만 가져오기
is_cross = df['light_origin'] == 1
df = df[is_cross]
df
```

```python
#상위 5%출력
is_cluster = df['total'] >= 0.9908
df_5 = df[is_cluster]
df_5
```

```python
#상위 10%출력
is_cluster = df['total'] >= 0.9842
df_10 = df[is_cluster]
df_10
```

```python
df_5.to_csv("cluster_0.05.csv")
df_10.to_csv("cluster_0.1.csv")
df_15.to_csv("cluster_0.15.csv")
```
