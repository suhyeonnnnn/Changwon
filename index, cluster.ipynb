```python
num =pd.read_csv('num_merge.csv')
num.head()
```

```python
df = pd.read_csv('changwon_300M_master_0927v04.csv')
df.head()
```

```python
df = pd.merge(df, num, how='left', on=('fid','id'))
```

```python
df.info()
```

```python
df['total'] = 0.02*df['bus'] + 0.32*df['dead']+ 0.16*df['aca'] + 0.15*df['kin'] +0.13*df['ele'] +0.12*df['num_10_fin'] +0.1*df['num_total_fin']
```

```python
df['total'].value_counts()
```

```python
#신호등이 있는 좌표만 가져오기
is_cross = df['light_origin'] == 1
df2 = df[is_cross]
df2
```
